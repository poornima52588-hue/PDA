```python
#Import libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import RobustScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#Load the DATA
df = pd.read_csv("/content/Shared_Engineered_Features (2).csv")

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nFirst 5 rows:")
print(df.head())
```

```python
#Convert date column and sort
df["week_start"] = pd.to_datetime(df["week_start"], errors="coerce")
df = df.sort_values("week_start").reset_index(drop=True)

print(df[["week_start"]].head())
print(df[["week_start"]].tail())
```

```python
#Check missing values
print(df.isnull().sum())
```

```python
#Define target and features
target_col = "total_resp"

drop_cols = [
    "week_start",
    "total_resp",
    "Adenovirus",
    "hMPV",
    "Influenza",
    "RSV"
]

feature_cols = [col for col in df.columns if col not in drop_cols]

print("Number of features:", len(feature_cols))
print(feature_cols)
```

```python
#Make sure boolean columns become numeric
for col in feature_cols:
    if df[col].dtype == "bool":
        df[col] = df[col].astype(int)
```

```python
#Keep only modeling columns and drop missing rows
model_df = df[["week_start"] + feature_cols + [target_col]].copy()
model_df = model_df.dropna().reset_index(drop=True)

print("Shape after dropping missing rows:", model_df.shape)
print(model_df.head())
```

```python
#Log-transform the target
model_df["log_target"] = np.log1p(model_df[target_col])
```

```python
#Quick EDA
#Summary statistics
print(model_df[feature_cols + [target_col]].describe())
#Correlation with target
corr_with_target = model_df[feature_cols + [target_col]].corr()[target_col].sort_values(ascending=False)
print(corr_with_target)
#Heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(model_df[feature_cols + [target_col]].corr(), cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
#Chronological 80:10:10 split
n = len(model_df)

train_end = int(n * 0.80)
val_end = int(n * 0.90)

train_df = model_df.iloc[:train_end].copy()
val_df = model_df.iloc[train_end:val_end].copy()
test_df = model_df.iloc[val_end:].copy()

print("Train shape:", train_df.shape)
print("Validation shape:", val_df.shape)
print("Test shape:", test_df.shape)
```

```python
#Split X and y
X_train = train_df[feature_cols]
X_val = val_df[feature_cols]
X_test = test_df[feature_cols]

y_train_log = train_df["log_target"]
y_val_log = val_df["log_target"]
y_test_log = test_df["log_target"]

y_train_actual = train_df[target_col].values
y_val_actual = val_df[target_col].values
y_test_actual = test_df[target_col].values
```

```python
#Scale the features
x_scaler = RobustScaler()

X_train_scaled = x_scaler.fit_transform(X_train)
X_val_scaled = x_scaler.transform(X_val)
X_test_scaled = x_scaler.transform(X_test)

print(X_train_scaled.shape, X_val_scaled.shape, X_test_scaled.shape)
```

```python
#Build the MLP model
mlp = MLPRegressor(
    hidden_layer_sizes=(32, 16),
    activation="relu",
    solver="adam",
    alpha=0.01,
    learning_rate="adaptive",
    learning_rate_init=0.001,
    early_stopping=True,
    validation_fraction=0.15,
    n_iter_no_change=20,
    max_iter=2000,
    random_state=42
)
```

```python
#Train the model
mlp.fit(X_train_scaled, y_train_log)

print("Iterations used:", mlp.n_iter_)
```

```python
#Predict
train_pred_log = mlp.predict(X_train_scaled)
val_pred_log = mlp.predict(X_val_scaled)
test_pred_log = mlp.predict(X_test_scaled)

train_pred = np.expm1(train_pred_log)
val_pred = np.expm1(val_pred_log)
test_pred = np.expm1(test_pred_log)

train_pred = np.maximum(train_pred, 0)
val_pred = np.maximum(val_pred, 0)
test_pred = np.maximum(test_pred, 0)
```

```python
#Evaluate the model
def evaluate_model(y_true, y_pred, name):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"\n{name} Performance")
    print(f"RMSE: {rmse:.4f}")
    print(f"MAE : {mae:.4f}")
    print(f"R²  : {r2:.4f}")

    return rmse, mae, r2

train_rmse, train_mae, train_r2 = evaluate_model(y_train_actual, train_pred, "Train")
val_rmse, val_mae, val_r2 = evaluate_model(y_val_actual, val_pred, "Validation")
test_rmse, test_mae, test_r2 = evaluate_model(y_test_actual, test_pred, "Test")
```

```python
#Put metrics into a table
metrics_df = pd.DataFrame({
    "Dataset": ["Train", "Validation", "Test"],
    "RMSE": [train_rmse, val_rmse, test_rmse],
    "MAE": [train_mae, val_mae, test_mae],
    "R2": [train_r2, val_r2, test_r2]
})

print(metrics_df)
```

```python
#Plot actual vs predicted on test set
plt.figure(figsize=(12, 5))
plt.plot(test_df["week_start"], y_test_actual, label="Actual", marker="o")
plt.plot(test_df["week_start"], test_pred, label="Predicted", marker="x")
plt.title("MLP: Test Set Actual vs Predicted")
plt.xlabel("Week Start")
plt.ylabel("Total Respiratory Cases")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Plot loss curve
plt.figure(figsize=(8, 5))
plt.plot(mlp.loss_curve_)
plt.title("MLP Training Loss Curve")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Save results
output_folder = "MLP_Shared_Engineered_Features_Output"
os.makedirs(output_folder, exist_ok=True)

metrics_df.to_csv(f"{output_folder}/mlp_metrics.csv", index=False)

pred_df = pd.DataFrame({
    "week_start": test_df["week_start"].values,
    "Actual_total_resp": y_test_actual,
    "Predicted_total_resp": test_pred
})
pred_df.to_csv(f"{output_folder}/mlp_test_predictions.csv", index=False)

plt.figure(figsize=(12, 5))
plt.plot(test_df["week_start"], y_test_actual, label="Actual", marker="o")
plt.plot(test_df["week_start"], test_pred, label="Predicted", marker="x")
plt.title("MLP: Test Set Actual vs Predicted")
plt.xlabel("Week Start")
plt.ylabel("Total Respiratory Cases")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{output_folder}/mlp_test_actual_vs_predicted.png", dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(mlp.loss_curve_)
plt.title("MLP Training Loss Curve")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{output_folder}/mlp_loss_curve.png", dpi=300)
plt.show()

print("Files saved in:", output_folder)
```
