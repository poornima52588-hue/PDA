```python
# 1. Install packages (run once)
# 2. Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# 3. Read processed datasets
weather = pd.read_csv("/content/drive/MyDrive/New Data/Processed_Weather.csv")
air = pd.read_csv("/content/drive/MyDrive/New Data/Processed_Air_Quality.csv")
traffic = pd.read_csv("/content/drive/MyDrive/New Data/Processed_Traffic.csv")
resp = pd.read_csv("/content/drive/MyDrive/New Data/Processed_Respiratory_Diseases.csv")
```

```python
# 4. Convert date columns
weather["time"] = pd.to_datetime(weather["time"])
air["time"] = pd.to_datetime(air["time"])
traffic["date"] = pd.to_datetime(traffic["date"])
resp["date"] = pd.to_datetime(resp["date"])
```

```python
# 5. Aggregate region-level daily data to national daily level
weather_daily = (
    weather.groupby("time", as_index=False)
    .agg({
        "temperature_2m_max": "mean",
        "temperature_2m_min": "mean",
        "precipitation_sum": "mean",
        "windspeed_10m_max": "mean"
    })
    .rename(columns={
        "temperature_2m_max": "Temp_Max",
        "temperature_2m_min": "Temp_Min",
        "precipitation_sum": "Precipitation",
        "windspeed_10m_max": "WindSpeed"
    })
)

air_daily = (
    air.groupby("time", as_index=False)
    .agg({
        "pm10": "mean",
        "pm2_5": "mean",
        "nitrogen_dioxide": "mean",
        "ozone": "mean",
        "sulphur_dioxide": "mean"
    })
    .rename(columns={
        "pm10": "PM10",
        "pm2_5": "PM2_5",
        "nitrogen_dioxide": "NO2",
        "ozone": "Ozone",
        "sulphur_dioxide": "SO2"
    })
)

traffic_daily = (
    traffic.groupby("date", as_index=False)
    .agg({
        "vehicle_count": "sum"
    })
    .rename(columns={
        "date": "time",
        "vehicle_count": "Traffic_Volume"
    })
)
```

```python
# 6. Merge daily predictors
daily_data = weather_daily.merge(air_daily, on="time", how="inner") \
                          .merge(traffic_daily, on="time", how="inner") \
                          .rename(columns={"time": "Date"})
```

```python
# 7. Create week start variable
daily_data["week_start"] = daily_data["Date"] - pd.to_timedelta(daily_data["Date"].dt.weekday, unit="D")
resp["week_start"] = resp["date"]
```

```python
# 8. Convert daily predictors to weekly predictors
weekly_predictors = (
    daily_data.groupby("week_start", as_index=False)
    .agg({
        "Temp_Max": "mean",
        "Temp_Min": "mean",
        "Precipitation": "mean",
        "WindSpeed": "mean",
        "PM10": "mean",
        "PM2_5": "mean",
        "NO2": "mean",
        "Ozone": "mean",
        "SO2": "mean",
        "Traffic_Volume": "mean"
    })
)
```

```python
# 9. Create target variable
resp["Respiratory_Burden"] = (
    resp["Adenovirus"] +
    resp["hMPV"] +
    resp["Influenza"] +
    resp["RSV"]
)
```

```python
# 10. Merge predictors with target
model_data = weekly_predictors.merge(
    resp[["week_start", "Adenovirus", "hMPV", "Influenza", "RSV", "Respiratory_Burden"]],
    on="week_start",
    how="inner"
)

model_data = model_data.sort_values("week_start").reset_index(drop=True)
model_data = model_data.dropna()
```

```python
# EXPLORATORY DATA ANALYSIS (EDA)
# 11. Basic data checks
print("Model Data Shape:")
print(model_data.shape)

print("\nColumn Names:")
print(model_data.columns.tolist())

print("\nSummary Statistics:")
print(model_data.describe())
```

```python
# 12. Correlation heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(
    model_data.drop(columns=["week_start"]).corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
# 13. Time-series plots
eda_plot = model_data.set_index("week_start")

eda_plot[[
    "Respiratory_Burden",
    "Temp_Max",
    "Temp_Min",
    "Precipitation",
    "WindSpeed",
    "PM10",
    "PM2_5",
    "NO2",
    "Ozone",
    "SO2",
    "Traffic_Volume"
]].plot(figsize=(14, 7))

plt.title("Weekly Time Series Trends")
plt.xlabel("Week")
plt.ylabel("Value")
plt.tight_layout()
plt.show()
```

```python
# 14. Histograms
model_data.drop(columns=["week_start"]).hist(figsize=(14, 10))
plt.suptitle("Distribution of Variables")
plt.tight_layout()
plt.show()
```

```python
# PRINCIPAL COMPONENT ANALYSIS (PCA)
# 15. Define predictor variables
feature_cols = [
    "Temp_Max", "Temp_Min", "Precipitation", "WindSpeed",
    "PM10", "PM2_5", "NO2", "Ozone", "SO2", "Traffic_Volume"
]

target_col = "Respiratory_Burden"

X = model_data[feature_cols]
y = model_data[target_col]
```

```python
# 16. Standardize predictors
scaler_pca = StandardScaler()
X_scaled_pca = scaler_pca.fit_transform(X)
```

```python
# 17. Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled_pca)

explained_variance = pca.explained_variance_ratio_

print("\nExplained Variance Ratio:")
for i, variance in enumerate(explained_variance, start=1):
    print(f"PC{i}: {variance:.4f}")
```

```python
# 18. Plot cumulative explained variance
plt.figure(figsize=(8, 5))
plt.plot(
    range(1, len(explained_variance) + 1),
    explained_variance.cumsum(),
    marker="o"
)
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Explained Variance")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 19. PCA loadings
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f"PC{i}" for i in range(1, len(feature_cols) + 1)],
    index=feature_cols
)

print("\nPCA Loadings:")
print(loadings)
```

```python
# LSTM DEEP LEARNING MODEL
# 20. Scale predictors and target
feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()

X_scaled = feature_scaler.fit_transform(model_data[feature_cols])
y_scaled = target_scaler.fit_transform(model_data[[target_col]])
```

```python
# 21. Create sequences
def create_sequences(X, y, window_size=4):
    X_seq, y_seq = [], []
    for i in range(window_size, len(X)):
        X_seq.append(X[i - window_size:i])
        y_seq.append(y[i])
    return np.array(X_seq), np.array(y_seq)

window_size = 4
X_seq, y_seq = create_sequences(X_scaled, y_scaled, window_size)
```

```python
# 22. Time-based train-test split 80% training, 20% testing
split_index = int(0.8 * len(X_seq))

X_train = X_seq[:split_index]
X_test = X_seq[split_index:]
y_train = y_seq[:split_index]
y_test = y_seq[split_index:]
```

```python
# 23. Build LSTM model
model = Sequential([
    LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1)
])
```

```python
# 24. Compile model
model.compile(
    optimizer="adam",
    loss="mse"
)
```

```python
# 25. Early stopping
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)
```

```python
# 26. Train model
history = model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=8,
    validation_data=(X_test, y_test),
    callbacks=[early_stop],
    verbose=1
)
```

```python
# 27. Make predictions
y_pred_scaled = model.predict(X_test)
```

```python
# 28. Inverse transform predictions
y_test_actual = target_scaler.inverse_transform(y_test)
y_pred_actual = target_scaler.inverse_transform(y_pred_scaled)
```

```python
# 29. Evaluate model performance
rmse_value = np.sqrt(mean_squared_error(y_test_actual, y_pred_actual))
mae_value = mean_absolute_error(y_test_actual, y_pred_actual)
r2_value = r2_score(y_test_actual, y_pred_actual)

print("\n====================================")
print("LSTM Model Performance")
print("====================================")
print("RMSE :", round(rmse_value, 3))
print("MAE  :", round(mae_value, 3))
print("R²   :", round(r2_value, 3))
```

```python
# 30. Save model metrics
metrics_df = pd.DataFrame({
    "Target": [target_col],
    "RMSE": [rmse_value],
    "MAE": [mae_value],
    "R2": [r2_value]
})

print("\nMetrics:")
print(metrics_df)

metrics_df.to_csv("LSTM_Model_Metrics.csv", index=False)
```

```python
# 31. Save prediction results
results = pd.DataFrame({
    "Week": model_data["week_start"].iloc[window_size + split_index:].values,
    "Actual": y_test_actual.flatten(),
    "Predicted": y_pred_actual.flatten()
})

results["Error"] = results["Actual"] - results["Predicted"]

print("\nPrediction Results Preview:")
print(results.head())

results.to_csv("LSTM_Weekly_Respiratory_Predictions.csv", index=False)
```

```python
# 32. Plot training and validation loss
plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("LSTM Training and Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 33. Actual vs Predicted scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(y_test_actual, y_pred_actual)
plt.xlabel(f"Actual {target_col}")
plt.ylabel(f"Predicted {target_col}")
plt.title(f"Actual vs Predicted - {target_col}")

min_val = min(y_test_actual.min(), y_pred_actual.min())
max_val = max(y_test_actual.max(), y_pred_actual.max())
plt.plot([min_val, max_val], [min_val, max_val])
plt.tight_layout()
plt.show()
```
