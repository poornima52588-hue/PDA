```python
#Import libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#Load the processed datasets
air = pd.read_csv("/content/Processed_Air_Quality.csv")
resp = pd.read_csv("/content/Processed_Respiratory_Diseases.csv")
traffic = pd.read_csv("/content/Processed_Traffic.csv")
weather = pd.read_csv("/content/Processed_Weather.csv")

print("Air shape:", air.shape)
print("Resp shape:", resp.shape)
print("Traffic shape:", traffic.shape)
print("Weather shape:", weather.shape)
```

```python
#Check columns
print("\nAir columns:")
print(air.columns.tolist())

print("\nResp columns:")
print(resp.columns.tolist())

print("\nTraffic columns:")
print(traffic.columns.tolist())

print("\nWeather columns:")
print(weather.columns.tolist())
```

```python
#Convert date columns
air["time"] = pd.to_datetime(air["time"], errors="coerce")
weather["time"] = pd.to_datetime(weather["time"], errors="coerce")
traffic["date"] = pd.to_datetime(traffic["date"], errors="coerce")
resp["date"] = pd.to_datetime(resp["date"], errors="coerce")
```

```python
#Create total respiratory target
resp["Total_Resp"] = (
    resp["Adenovirus"] +
    resp["hMPV"] +
    resp["Influenza"] +
    resp["RSV"]
)

print(resp[["date", "Adenovirus", "hMPV", "Influenza", "RSV", "Total_Resp"]].head())
```

```python
#Aggregate air quality to daily national averages
air_daily = air.groupby("time", as_index=False)[
    ["pm2_5", "nitrogen_dioxide", "ozone"]
].mean()

air_daily = air_daily.rename(columns={"time": "date"})
print(air_daily.head())
```

```python
#Aggregate weather to daily national averages
weather_daily = weather.groupby("time", as_index=False)[
    ["temperature_2m_max"]
].mean()

weather_daily = weather_daily.rename(columns={"time": "date"})
print(weather_daily.head())
```

```python
#Aggregate traffic to daily national averages
traffic_daily = traffic.groupby("date", as_index=False)[
    ["vehicle_count"]
].mean()

print(traffic_daily.head())
```

```python
#Merge daily external datasets
daily_df = air_daily.merge(weather_daily, on="date", how="inner")
daily_df = daily_df.merge(traffic_daily, on="date", how="inner")

daily_df = daily_df.sort_values("date").reset_index(drop=True)

print("Daily merged shape:", daily_df.shape)
print(daily_df.head())
```

```python
#Create daily lag and rolling features
daily_df["temperature_2m_max_lag_4d"] = daily_df["temperature_2m_max"].shift(4)
daily_df["nitrogen_dioxide_lag_4d"] = daily_df["nitrogen_dioxide"].shift(4)
daily_df["pm2_5_lag_7d"] = daily_df["pm2_5"].shift(7)

daily_df["temperature_2m_max_rolling_7d"] = daily_df["temperature_2m_max"].rolling(7).mean()
daily_df["vehicle_count_rolling_7d"] = daily_df["vehicle_count"].rolling(7).mean()

daily_df["ozone_rolling_14d"] = daily_df["ozone"].rolling(14).mean()
daily_df["temperature_2m_max_rolling_14d"] = daily_df["temperature_2m_max"].rolling(14).mean()
daily_df["nitrogen_dioxide_rolling_14d"] = daily_df["nitrogen_dioxide"].rolling(14).mean()
```

```python
#Convert daily external data to weekly
daily_df = daily_df.set_index("date")
weekly_external = daily_df.resample("W-MON").mean().reset_index()

print("Weekly external shape:", weekly_external.shape)
print(weekly_external.head())
```

```python
#Keep weekly respiratory data
resp_weekly = resp[["date", "Total_Resp"]].copy()
resp_weekly = resp_weekly.sort_values("date").reset_index(drop=True)

print(resp_weekly.head())
```

```python
#Merge weekly target with weekly external features
df = resp_weekly.merge(weekly_external, on="date", how="inner")
df = df.sort_values("date").reset_index(drop=True)

print("Merged weekly dataset shape:", df.shape)
print(df.head())
```

```python
#Create target lag features
df["Total_Resp_lag_1w"] = df["Total_Resp"].shift(1)
df["Total_Resp_lag_2w"] = df["Total_Resp"].shift(2)

print(df[["date", "Total_Resp", "Total_Resp_lag_1w", "Total_Resp_lag_2w"]].head(10))
```

```python
#Keep the curated feature set
feature_cols = [
    "Total_Resp_lag_1w",
    "Total_Resp_lag_2w",
    "ozone_rolling_14d",
    "temperature_2m_max_rolling_7d",
    "temperature_2m_max_rolling_14d",
    "temperature_2m_max_lag_4d",
    "nitrogen_dioxide_rolling_14d",
    "nitrogen_dioxide_lag_4d",
    "vehicle_count",
    "vehicle_count_rolling_7d",
    "pm2_5_lag_7d"
]

target_col = "Total_Resp"

model_df = df[["date"] + feature_cols + [target_col]].copy()
model_df = model_df.dropna().reset_index(drop=True)

print("Shape after dropping missing rows:", model_df.shape)
print(model_df.head())
```

```python
#Log-transform traffic features and target
model_df["vehicle_count"] = np.log1p(model_df["vehicle_count"])
model_df["vehicle_count_rolling_7d"] = np.log1p(model_df["vehicle_count_rolling_7d"])
model_df["log_target"] = np.log1p(model_df[target_col])
```

```python
#EDA
print(model_df[feature_cols + [target_col]].describe())
plt.figure(figsize=(10, 7))
sns.heatmap(model_df[feature_cols + [target_col]].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
#Scale the feature columns
x_scaler = RobustScaler()

X_all = x_scaler.fit_transform(model_df[feature_cols])
y_all = model_df["log_target"].values

print("X shape:", X_all.shape)
print("y shape:", y_all.shape)
```

```python
#Create time sequences
#For Transformer, we shall not use single rows only. We are creating a rolling window of past weeks.
#using 4 past weeks to predict the next week.
sequence_length = 4

def create_sequences(X, y, dates, seq_len):
    X_seq, y_seq, date_seq = [], [], []
    for i in range(seq_len, len(X)):
        X_seq.append(X[i-seq_len:i])
        y_seq.append(y[i])
        date_seq.append(dates.iloc[i])
    return np.array(X_seq), np.array(y_seq), np.array(date_seq)

X_seq, y_seq, date_seq = create_sequences(
    X_all,
    y_all,
    model_df["date"],
    sequence_length
)

print("Sequence X shape:", X_seq.shape)
print("Sequence y shape:", y_seq.shape)
```

```python
#Chronological 80:10:10 split
n = len(X_seq)

train_end = int(n * 0.80)
val_end = int(n * 0.90)

X_train = X_seq[:train_end]
X_val = X_seq[train_end:val_end]
X_test = X_seq[val_end:]

y_train = y_seq[:train_end]
y_val = y_seq[train_end:val_end]
y_test = y_seq[val_end:]

date_train = date_seq[:train_end]
date_val = date_seq[train_end:val_end]
date_test = date_seq[val_end:]

print("Train shape:", X_train.shape)
print("Validation shape:", X_val.shape)
print("Test shape:", X_test.shape)
```

```python
#Keep actual target values for final evaluation
actual_target = model_df[target_col].values[sequence_length:]

y_train_actual = actual_target[:train_end]
y_val_actual = actual_target[train_end:val_end]
y_test_actual = actual_target[val_end:]
```

```python
#Build Transformer block
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0.1):
    x = layers.MultiHeadAttention(
        key_dim=head_size,
        num_heads=num_heads,
        dropout=dropout
    )(inputs, inputs)

    x = layers.Dropout(dropout)(x)
    x = layers.LayerNormalization(epsilon=1e-6)(x + inputs)

    ffn = layers.Dense(ff_dim, activation="relu")(x)
    ffn = layers.Dropout(dropout)(ffn)
    ffn = layers.Dense(inputs.shape[-1])(ffn)

    return layers.LayerNormalization(epsilon=1e-6)(x + ffn)
```

```python
#Build the full Transformer model
input_shape = X_train.shape[1:]

inputs = layers.Input(shape=input_shape)

x = transformer_encoder(inputs, head_size=16, num_heads=2, ff_dim=32, dropout=0.1)
x = transformer_encoder(x, head_size=16, num_heads=2, ff_dim=32, dropout=0.1)

x = layers.GlobalAveragePooling1D()(x)
x = layers.Dense(32, activation="relu")(x)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(1)(x)

model = models.Model(inputs, outputs)

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="mse",
    metrics=["mae"]
)

model.summary()
```

```python
#Train the Transformer model
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=200,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)
```

```python
#Predict
train_pred_log = model.predict(X_train).flatten()
val_pred_log = model.predict(X_val).flatten()
test_pred_log = model.predict(X_test).flatten()

train_pred = np.maximum(np.expm1(train_pred_log), 0)
val_pred = np.maximum(np.expm1(val_pred_log), 0)
test_pred = np.maximum(np.expm1(test_pred_log), 0)
```

```python
#Evaluate the model
def evaluate_model(y_true, y_pred, name):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"\n{name} Performance")
    print(f"RMSE: {rmse:.4f}")
    print(f"MAE : {mae:.4f}")
    print(f"R²  : {r2:.4f}")

    return rmse, mae, r2

train_rmse, train_mae, train_r2 = evaluate_model(y_train_actual, train_pred, "Train")
val_rmse, val_mae, val_r2 = evaluate_model(y_val_actual, val_pred, "Validation")
test_rmse, test_mae, test_r2 = evaluate_model(y_test_actual, test_pred, "Test")
```

```python
#Put results into a table
metrics_df = pd.DataFrame({
    "Dataset": ["Train", "Validation", "Test"],
    "RMSE": [train_rmse, val_rmse, test_rmse],
    "MAE": [train_mae, val_mae, test_mae],
    "R2": [train_r2, val_r2, test_r2]
})

print(metrics_df)
```

```python
#Plot actual vs predicted on the test set
plt.figure(figsize=(12, 5))
plt.plot(date_test, y_test_actual, label="Actual", marker="o")
plt.plot(date_test, test_pred, label="Predicted", marker="x")
plt.title("Transformer: Test Set Actual vs Predicted")
plt.xlabel("Date")
plt.ylabel("Total_Resp")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Plot training history
plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Transformer Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Save results
output_folder = "Transformer_From_Processed_Data"
os.makedirs(output_folder, exist_ok=True)

metrics_df.to_csv(f"{output_folder}/transformer_metrics.csv", index=False)

pred_df = pd.DataFrame({
    "date": date_test,
    "Actual_Total_Resp": y_test_actual,
    "Predicted_Total_Resp": test_pred
})
pred_df.to_csv(f"{output_folder}/transformer_test_predictions.csv", index=False)

plt.figure(figsize=(12, 5))
plt.plot(date_test, y_test_actual, label="Actual", marker="o")
plt.plot(date_test, test_pred, label="Predicted", marker="x")
plt.title("Transformer: Test Set Actual vs Predicted")
plt.xlabel("Date")
plt.ylabel("Total_Resp")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{output_folder}/transformer_test_actual_vs_predicted.png", dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Transformer Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{output_folder}/transformer_training_loss.png", dpi=300)
plt.show()

print("Files saved in:", output_folder)
```
